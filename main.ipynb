```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

#you know wht I mean by this
```

```python
img = cv2.imread('Akshay.webp')

#read the image and the system already know who this guy is

img.shape
#first 2 values of this threee dimensional array here represents height and width of the image respectively 
#and 3rd value is the color channel which is often RGB,BGR and maybe sometimes Grayscale
```

```python
face_classifier = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
#firing up the classifier from cv library notice the xml file in parameters, it,s just a normal xml file containing the data of the model which was originally trained for. 
#It works like a charm for every face on the planet
```

```python
face = face_classifier.detectMultiScale(img,scaleFactor=1.1,minNeighbors=5,minSize=(40,40))
#read the word file for better clarification of this multiscale method and their parameters
```

```python
for (x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 255), 4)

#the almighty boundry box code finally here
    #img - image ðŸ‘€
    #x - starting point
    #y - ending point
    #h - height of box
    #w - width of box
    #(,,,) - color code
    #last value is thickness
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,10))
plt.imshow(img_rgb)
plt.axis('off')
```
